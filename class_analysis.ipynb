```python
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path
import numpy as np
import warnings
from copy import deepcopy

warnings.simplefilter(action='ignore', category=FutureWarning)
warnings.simplefilter(action='ignore', category=UserWarning)

import constants
import data
import results_gathering
import ml_importance_methods
```

```python
chemical_data = data.prepare_chemical_data()
chemical_data
```

```python
class_data, indices_to_names = data.prepare_class_data()
class_data
```

```python
decition_tree_importance = results_gathering.important_relationships_for_all_bacteria(
    chemical_data,
    class_data,
    ml_importance_methods.get_decision_tree_importance,
    model_name="decision_tree",
    importance_threshold=0.9)
results_to_display_decision_tree = deepcopy(decition_tree_importance)
names = results_gathering.name_lookup(list(results_to_display_decision_tree['Bacteria']), indices_to_names)
results_to_display_decision_tree['Bacteria'] = names
results_to_display_decision_tree
```

```python
sorted_decition_tree_importance = decition_tree_importance.sort_values("R_squared_decision_tree", ascending=False)
sorted_decition_tree_importance
```

```python
xgboost_feature_importance = results_gathering.important_relationships_for_all_bacteria(
    chemical_data,
    class_data,
    ml_importance_methods.get_xgboost_importance,
    model_name="xgboost",
    importance_threshold=0.9)
results_to_display_xgboost = deepcopy(xgboost_feature_importance)
names = results_gathering.name_lookup(list(results_to_display_xgboost['Bacteria']), indices_to_names)
results_to_display_xgboost['Bacteria'] = names
results_to_display_xgboost
```

```python
random_forest_feature_importance = results_gathering.important_relationships_for_all_bacteria(
    chemical_data,
    class_data,
    ml_importance_methods.get_random_forest_importance,
    model_name="random_forest",
    importance_threshold=0.15)
results_to_display_rf = deepcopy(random_forest_feature_importance)
names = results_gathering.name_lookup(list(results_to_display_rf['Bacteria']), indices_to_names)
results_to_display_rf['Bacteria'] = names
results_to_display_rf
```

```python
final_results = results_gathering.prepare_final_results(
    chemical_data,
    class_data,
    decition_tree_importance,
    xgboost_feature_importance,
    random_forest_feature_importance,
)
names = results_gathering.name_lookup(list(final_results['Bacteria']), indices_to_names)
final_results['Bacteria'] = names
final_results
```

```python
decition_tree_importance.to_csv("./results/class_analysis_decision_tree.csv")
xgboost_feature_importance.to_csv("./results/class_analysis_xgboost.csv")
random_forest_feature_importance.to_csv("./results/class_analysis_rf.csv")
```
