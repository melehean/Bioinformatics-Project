```python
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
warnings.simplefilter(action='ignore', category=UserWarning)

import constants
import data
import results_gathering
import ml_importance_methods
```

```python
chemical_data = data.prepare_chemical_data()
chemical_data
```

```python
family_data, family_names = data.prepare_family_data(drop_unknown_families=False)#data.prepare_family_data(True)
family_data
```

```python
decition_tree_importance = results_gathering.important_relationships_for_all_bacteria(
    chemical_data,
    family_data,
    ml_importance_methods.get_decision_tree_importance,
    model_name="decision_tree",
    importance_threshold=0.9)
#decition_tree_importance

names = results_gathering.name_lookup(list(decition_tree_importance['Bacteria']), family_names)
decition_tree_importance_with_names=decition_tree_importance.copy()
decition_tree_importance_with_names['Bacteria'] = names
decition_tree_importance_with_names
```

```python
decition_tree_importance.to_csv("./results/decition_tree_importance_family.csv")
```

```python
xgboost_feature_importance = results_gathering.important_relationships_for_all_bacteria(
    chemical_data,
    family_data,
    ml_importance_methods.get_xgboost_importance,
    model_name="xgboost",
    importance_threshold=1.0)
#xgboost_feature_importance

names = results_gathering.name_lookup(list(xgboost_feature_importance['Bacteria']), family_names)
xgboost_feature_importance_with_names=xgboost_feature_importance.copy()
xgboost_feature_importance_with_names['Bacteria'] = names
xgboost_feature_importance_with_names
```

```python
xgboost_feature_importance.to_csv("./results/xgboost_feature_importance_family.csv")
```

```python
sorted_xgboost_feature_importance_with_names = xgboost_feature_importance_with_names.sort_values("CV_xgboost", ascending=False)
sorted_xgboost_feature_importance_with_names

#names = results_gathering.name_lookup(list(sorted_xgboost_feature_importance['Bacteria']), family_names)
#sorted_xgboost_feature_importance['Bacteria'] = names
#sorted_xgboost_feature_importance
```

```python
sorted_xgboost_feature_importance_with_names.to_csv("./results/sorted_xgboost_feature_importance_family.csv")
```

```python
random_forest_feature_importance = results_gathering.important_relationships_for_all_bacteria(
    chemical_data,
    family_data,
    ml_importance_methods.get_random_forest_importance,
    model_name="random_forest",
    importance_threshold=0.19)
names = results_gathering.name_lookup(list(random_forest_feature_importance['Bacteria']), family_names)
random_forest_feature_importance['Bacteria'] = names
random_forest_feature_importance
```

```python
random_forest_feature_importance.to_csv("./results/random_forest_feature_importance_family.csv")
```

```python
sorted_rf_feature_importance = random_forest_feature_importance.sort_values("CV_random_forest", ascending=False)
sorted_rf_feature_importance
```

```python
sorted_rf_feature_importance.to_csv("./results/sorted_rf_feature_importance_family.csv")
```
